```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import correlate, convolve

%load_ext autoreload
%autoreload 2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)
plt.set_cmap('Greys_r')
```

```python
# Load and display original image
pollen = np.array(Image.open('./Pollen.png')).astype(np.float32) / 255.
print(pollen.shape)
plt.imshow(pollen)
plt.show()
```

```python
filt_shape = (5, 5)
filt = np.ones(filt_shape) / np.prod(filt_shape)
```

```python
# Blurring via linear filtering
pollen_lin_filt = correlate(pollen, filt)
plt.imshow(pollen_lin_filt)
plt.show()
```

```python
# Edge detection
pollen_subt = pollen - pollen_lin_filt
plt.imshow(pollen_subt, vmin=0., vmax=1.)
plt.show()
```

```python
pollen_add = pollen + pollen_subt
plt.imshow(pollen_add, vmin=0., vmax=1.)
plt.show()
```

```python
filt20_shape = (20, 20)
filt20 = np.ones(filt20_shape) / np.prod(filt20_shape)
pollen_lin_filt20 = correlate(pollen, filt20)
plt.imshow(pollen_lin_filt20, vmin=0., vmax=1.)
```

```python
pollen_subt20 = pollen - pollen_lin_filt20
plt.imshow(pollen_subt20, vmin=0., vmax=1.)
plt.show()
```

```python
pollen_add = pollen + pollen_subt20
plt.imshow(pollen_add, vmin=0., vmax=1.)
plt.show()
```
