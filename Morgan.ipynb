```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage.filters import convolve
from fspecial import fspecial

%load_ext autoreload
%autoreload 2

plt.set_cmap('Greys_r')
```

```python
# Load image and convert to numpy nd-array
morgan = np.array(Image.open('./Morgan.tif'))
plt.imshow(morgan)
plt.show()
```

```python
r = 5
X, Y = np.meshgrid(-r:r, -r:r)
```

```python
disk = fspecial('disk', radius=2)
morgan_disk = convolve(morgan, disk)
plt.imshow(morgan_disk)
plt.show()
```

```python
avg = fspecial('avg', fsize=(5, 5))
morgan_avg = convolve(morgan, avg)
plt.imshow(morgan_avg)
plt.show()
```

```python
gaussian = fspecial('gaussian', shape=(5, 5))
morgan_gauss = convolve(morgan, gaussian)
plt.imshow(morgan_gauss)
plt.show()
```

```python
unsharp = fspecial('unsharp')
morgan_unsharp = convolve(morgan, unsharp)
plt.imshow(morgan_unsharp)
plt.show()
```
